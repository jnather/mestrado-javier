```python
import numpy as np

# media de idade de todos 



# Group by 'COD_PACIENTE' and calculate age for each unique patient
unique_patients = df_.groupby('COD_PACIENTE').first()

print("Number unique COD_REGISTRO: ", len(unique_patients))

# Calculate the age in years
unique_patients['AGE'] = ((unique_patients['DTA_EXAME'] - unique_patients['DTA_NASCIMENTO']) / np.timedelta64(1, 'D')) / 365.25

# Calculate the mean age of all unique patients
mean_age = unique_patients['AGE'].mean()

# Calculate the variance of ages for all unique patients
variance_age = unique_patients['AGE'].var()

print("Mean age of unique patients:", mean_age)
print("Variance of ages for unique patients:", variance_age)

# Calculate the minimum and maximum ages of all unique patients
min_age = unique_patients['AGE'].min()
max_age = unique_patients['AGE'].max()

print("Minimum age of unique patients:", min_age)
print("Maximum age of unique patients:", max_age)

# Count the number of patients less than 18 years old
under_18_count = len(unique_patients[unique_patients['AGE'] < 18])

# Count the number of patients 18 years or older
above_or_equal_18_count = len(unique_patients[unique_patients['AGE'] >= 18])

print("Number of patients under 18 years old:", under_18_count)
print("Number of patients 18 years or older:", above_or_equal_18_count)

# Count the number of male patients
male_count = len(unique_patients[unique_patients['SEXO'] == 'M'])

# Count the number of female patients
female_count = len(unique_patients[unique_patients['SEXO'] == 'F'])

print("Number of male patients:", male_count)
print("Number of female patients:", female_count)

```

```python
def calcular_estatisticas(df_junia):
    # Calculate the mean age of all unique patients
    mean_age = df_junia['IDADE'].mean()

    print("Mean age of unique patients:", mean_age)

    # Calculate the minimum and maximum ages of all unique patients
    min_age = df_junia['IDADE'].min()
    max_age = df_junia['IDADE'].max()

    print("Minimum age of unique patients:", min_age)
    print("Maximum age of unique patients:", max_age)

    # Count the number of patients less than 18 years old
    under_18_count = len(df_junia[df_junia['IDADE'] < 18])

    # Count the number of patients 18 years or older
    above_or_equal_18_count = len(df_junia[df_junia['IDADE'] >= 18])

    # Count the number of patients 60 years or older
    above_60_count = len(df_junia[df_junia['IDADE'] > 60])


    print("Number of patients under 18 years old:", under_18_count)
    print("Number of patients 18 years or older:", above_or_equal_18_count)
    print("Number of patients 60 years or older:", above_60_count)

    # Count the number of male patients
    male_count = len(df_junia[df_junia['IDF_SEXO'] == 'M'])

    # Count the number of female patients
    female_count = len(df_junia[df_junia['IDF_SEXO'] == 'F'])

    print("Number of male patients:", male_count)
    print("Number of female patients:", female_count)

    # Raca
    print(" --- \n","Contagem de cada raça: ", df_junia["RACA"].value_counts(), "\n --- ")

    # alta internacao
    print(" --- \n",df_junia["TIPO_SAIDA_ATEND"].value_counts(), "\n --- ")

    # Internacao
    print(" --- \n",df_junia["INTERNADO"].value_counts(), "\n --- ")

    
```
